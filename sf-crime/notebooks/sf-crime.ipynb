```python
import numpy as np          
import pandas as pd         
import matplotlib.pyplot as plt  
import seaborn as sns  
from pathlib import Path
```

```python
sfcrime2018 = pd.read_csv("/Users/drewvetteth/Summer-2025/venv/Incident_Reports_2018") 
sfcrimePresent = pd.read_csv("/Users/drewvetteth/Summer-2025/venv/Incident_Reports_Present")
sfcrime2018.head()
sfcrimePresent.head()
```

```python
simple_sfcrime2018 = sfcrime2018[["Date", "Time", "DayOfWeek", "PdDistrict", "Category","Descript","Resolution"]].sort_values(["Date", "Time"], ascending = True)
simple_sfcrimePresent = sfcrimePresent[["Incident Date", "Incident Time", "Incident Day of Week", "Police District", "Incident Category","Incident Description","Resolution"]]
simple_sfcrimePresent = simple_sfcrimePresent.rename(columns = {"Incident Date": "Date", "Incident Time": "Time", "Incident Day of Week" : "DayOfWeek", "Police District" : "PdDistrict", "Incident Category" : "Category", "Incident Description" : "Descript"})
simple_sfcrimePresent["Date"] = pd.to_datetime(simple_sfcrimePresent["Date"])
simple_sfcrimePresent["Date"] = simple_sfcrimePresent["Date"].dt.strftime("%m/%d/%Y")
simple_sfcrimePresent.sort_values(["Date", "Time"], ascending = True)
```

```python
print(simple_sfcrime2018.dtypes)
print(simple_sfcrimePresent.dtypes)
```

```python
all_crime = pd.concat([simple_sfcrime2018, simple_sfcrimePresent], ignore_index=True)
all_crime.sort_values(["Date", "Time"], ascending=True)
```

```python
precovid_crime = all_crime[all_crime["Date"] < "03/17/2020"]
covid_crime = all_crime[(all_crime["Date"] > "03/17/2020") & (all_crime["Date"] < "03/17/2022")]
postcovid_crime = all_crime[all_crime["Date"] > "03/17/2022"]
recent_crime = all_crime[all_crime["Date"] > "07/01/2019"]
```

```python
path = Path("/Users/drewvetteth/Summer-2025/sf-crime/data/Incident_Reports_Present") 

dates = pd.read_csv(path, parse_dates=["Incident Datetime"])

dates["year"] = dates["Incident Datetime"].dt.year
dates["half year"] = (dates["Incident Datetime"].dt.month > 6).astype(int) + 1

outdir = Path("/Users/drewvetteth/Summer-2025/sf-crime/data/sixmonths")
outdir.mkdir(parents=True, exist_ok=True)


for (y, h), g in dates.groupby(["year","half year"], sort=True):
    if ((y > 2019) | (y == 2019 & h == 2)):
        g.to_csv(outdir / f"sfcrime_{y}_H{h}.csv", index=False)

print("Done. Wrote:", sorted(p.name for p in outdir.glob("*.csv")))
```
